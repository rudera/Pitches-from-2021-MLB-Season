```python
from baseball_scraper import statcast
data = statcast(start_dt='2021-04-01',end_dt='2021-10-03')
```

```python
data.columns
```

```python
data.info
```

```python
data.to_csv('MLB2021_Pitches.csv')
```

```python
import seaborn as sns
import pandas as pd
```

```python
pitches = data.groupby(data['pitch_type'])
pitch_types = pitches.count()['index']
pitch_types
```

```python
sns.barplot(pitch_types.index, pitch_types.values)
```

```python
pitchers = data.groupby(data['pitch_type']).get_group('FF').groupby(data['player_name']).mean()['release_speed']
pitchers
```

```python
pitch_count = data.groupby(data['pitch_type']).get_group('FF').groupby(data['player_name']).count()['index']
pitch_count.sort_values()
```

```python
x, y = pd.Series(pitch_count, name="Fastballs Thrown"), pd.Series(pitchers, name="Average  Velocity")
sns.regplot(x=x,y=y)
```

```python
data.groupby(data['events']).get_group('double')['launch_speed'].sort_values()
```

```python
shome = data.loc[data['batter']==660271].groupby('events').get_group('home_run')[['pitch_type','release_speed','launch_angle','launch_speed']]
shome
```
